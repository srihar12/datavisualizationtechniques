```python
import pandas as pd
import sqlite3


df = pd.read_csv(r"C:\Users\HDC0422062\Downloads\tourism_dataset.csv")


conn = sqlite3.connect("tourism.db")


df.to_sql("tourism", conn, if_exists="replace", index=False)


query = "SELECT * FROM tourism"

sql_df = pd.read_sql(query, conn)

print(sql_df)


print(sql_df.head())

conn.close()
```
