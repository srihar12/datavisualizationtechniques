```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\kalap\Downloads\tourism_dataset.csv")
print(df)
df.head()
df.info()
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
Q1 = df["Revenue"].quantile(0.25)

Q3 = df["Revenue"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR

upper = Q3 + 1.5 * IQR

outliers = df[(df["Revenue"] < lower) |
              (df["Revenue"] > upper)]

print(outliers)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[["Visitors","Rating","Revenue"]] = scaler.fit_transform(
    df[["Visitors","Rating","Revenue"]]
)

print(df[["Visitors","Rating","Revenue"]])
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df["Country"] = encoder.fit_transform(df["Country"])

df["Category"] = encoder.fit_transform(df["Category"])

df["Accommodation_Available"] = encoder.fit_transform(df["Accommodation_Available"])

print(df.head())
```

```python
encoded = pd.get_dummies(
    df,
    columns=["Country","Category","Accommodation_Available"]
)

print(encoded.head())
df.to_csv("cleaned_tourism_dataset.csv", index=False)
```
